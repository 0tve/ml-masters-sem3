```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


import warnings
warnings.filterwarnings('ignore')
```

```python
X = np.array([
    [50, 2],   # Площадь и количество комнат
    [70, 3],
    [100, 4],
    [120, 5],
    [80, 3],
    [60, 2],
    [150, 5],
    [90, 3],
    [110, 4],
    [130, 5],
    [140, 5],
    [45, 1],
    [55, 2],
    [85, 3],
    [95, 3],
    [75, 3],
    [65, 2],
    [125, 4],
    [135, 5],
    [160, 6]
])

y = np.array([
    150, 200, 300, 400, 250, 180, 500, 270, 350, 450, 480,
    120, 160, 260, 290, 230, 190, 370, 460, 600
])  # Цены домов
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train = scaler_y.fit_transform(y_train.reshape(-1, 1))
y_test = scaler_y.transform(y_test.reshape(-1, 1))
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),  # Первый слой: 8 нейронов, ReLU
    Dense(4, activation='relu'),                   # Второй слой: 4 нейрона, ReLU
    Dense(1, activation='linear')                  # Выходной слой: 1 нейрон, линейная активация
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error', 'mean_absolute_error'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1)
```

```python
loss, mse, mae = model.evaluate(X_test, y_test)
print(f"Среднеквадратичная ошибка на тестовой выборке: {mse:.2f}")
print(f"Абсолютная квадратичная ошибка на тестовой выборке: {mae:.2f}")
```

```python
new_data = np.array([[80, 3]])  # Новый пример
new_data_scaled = scaler_X.transform(new_data)
predicted_price_scaled = model.predict(new_data_scaled)
predicted_price = scaler_y.inverse_transform(predicted_price_scaled)
print(f"Прогнозируемая цена дома: ${predicted_price[0][0]:.2f} тыс.")
```
